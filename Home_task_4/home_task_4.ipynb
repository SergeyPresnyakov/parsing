```python
with open('result.json') as f:
    vacancy = json.load(f)
```

```python
# Посмотрим на вакансии
vacancy
```

```python
for item in vacancy:
   try:
      id = {"_id" : hash(json.dumps(item))}
      collection.insert_one(dict(list(id.items()) + list(item.items())))
   except DuplicateKeyError:
      print('Объект с _id:\t%s уже записан в колекцию'%id['_id'])
```

2. Написать функцию, которая производит поиск и выводит на экран вакансии с заработной платой больше введённой суммы (необходимо анализировать оба поля зарплаты).

```python
def searh_salary(salary):
    for doc in  collection.find({'$or': [{'salary_min': {'$gte': salary}}, {'salary_max': {'$gte': salary}}]}):
        pprint(doc)
```

```python
searh_salary(50000)
```

3. Любая аналитика. Например matching ваканский с разных площадок.

Я дополню предыдущую функцию возможностью выбора валюты. Вводить как в документах, например, RUB или USD

```python
def searh_salary_2(salary, currency):
    for doc in  collection.find({'$and': [{'$or': [{'salary_min': {'$gte': salary}}, {'salary_max': {'$gte': salary}}]}, {'currency': currency}]}):
        pprint(doc)
```

```python
searh_salary_2(50000, 'RUB')
```
