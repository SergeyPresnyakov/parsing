Структура данных в виде словаря должна содержать:
- *название источника;
- наименование новости;
- ссылку на новость;
- дата публикации.

Минимум один сайт, максимум - все два

```python
from pprint import pprint
from lxml import html
import requests
import pandas as pd
```

```python
url = 'https://lenta.ru/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/105.0.0.0 Safari/537.36',
}

params = {
}
```

```python
response = requests.get(url=url, headers=headers, params=params)
response.status_code, response.url
```

```python
dom = html.fromstring(response.text)
```

```python
# Все блоки новостей
news = dom.xpath('//a[contains(@class, "card")]')
```

```python
news_list = []
new_dict = {}
i = 0
for new in news:
    i = i + 1
    if len(new.xpath(".//*[name()='svg']/*[name()='use']/attribute::*")) > 0:
        if 'label' in new.xpath(".//*[name()='svg']/*[name()='use']/attribute::*")[0]:
          source = new.xpath(".//*[name()='svg']/*[name()='use']/attribute::*")[0][10:]
    else:
        source = 'Lenta.ru'
    
    name_news = (new.xpath(".//h3/text()") + new.xpath(".//span/text()"))[0]

    if len(new.xpath(".//time/text()")) > 0:
        date = new.xpath(".//time/text()")[0]
    else:
        date = None

    if len(new.xpath(".//@href")) > 0:
        link = 'https://lenta.ru' + new.xpath(".//@href")[0]
    else:
        link = None
    
    new_dict = {
      '№': i,
      'source': source,
      'name_news': name_news,
      'link': link,
      'time/date': date
      }
    news_list.append(new_dict)
```
